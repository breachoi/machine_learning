## 4. Methodology
### Data Preprocessing:
- Overview
The dataset contains a variety of columns with job-related information, such as job titles, industries, salary estimates, and more. The purpose of the preprocessing was to clean the data, handle missing values, and consolidate redundant information to prepare it for further analysis.

- Handling Missing Values
Upon inspection, several columns were identified to contain missing values. Specifically, the value "-1" in the Headquarters, Industry, and Sector columns was treated as missing data, as it indicates unavailable or inaccessible information. These missing values were imputed with relevant data. For the Size column, the value "unknown" was found, and since it represents an unknown but valid entry, we decided to retain it as it was.
In the case of the Competitors column, over 60% of its values were marked as "unknown", which we considered as missing data. Given the significant portion of missing data, we decided to drop this column as it would not significantly affect the analysis. Similarly, the Seniority column also contained more than 60% missing values, and therefore, was dropped as well, as it did not add substantial value to the analysis.
For the Job Title Sim column, since this column could be merged with the Job Title column, we chose to retain the missing values and handle them during the data preprocessing or merging phase. Lastly, for the Degree column, we decided to retain the "na" values, as they provide meaningful information, specifically indicating individuals who do not have a degree.

- Data Imputation
We proceeded to impute missing values for the Headquarters, Industry, and Sector columns, as these were deemed to contain valid information that could be filled based on the available data. For columns such as Competitors and Seniority, which had a significant amount of missing data, we opted to drop them, as they did not provide enough relevant data for further analysis.

- Data Standardization
The Job Title column contained inconsistencies, with various titles referring to similar roles. To standardize the data, we consolidated similar job titles into broader categories such as "Data Scientist," "Data Engineer," and "Data Analyst." This made the column more consistent and easier to analyze.
Additionally, we combined the Location and Job Location columns into a single new column called Combined Location. This eliminated redundancy by ensuring that when the Job Location matched the last part of the Location, the data was merged appropriately.

- Salary Estimate Cleaning
The Salary Estimate column contained extraneous symbols like dollar signs ("$") and the letter K, which needed to be removed to facilitate accurate analysis. We stripped these characters and then extracted the numeric values from salary ranges. We further calculated the mean of the lower and upper bounds in the salary range, providing a more consistent and standardized representation of salary data.

- Quartile Calculation
To better understand the distribution of salaries, we divided the Avg Salary (K) values into four quartiles using the qcut function. We then applied labels to the salary quartiles, ensuring that the data could be easily grouped into these categories for further analysis.

- Column Dropping
Several columns were dropped after the merging and cleaning processes. Specifically, the Job Title Sim, Seniority, Competitors, and the original columns that were combined to create new columns were dropped. These columns either contained too much missing data or did not provide substantial value to the analysis.

```python
#Necessary Library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import re
import seaborn as sns
```

```python
#Load the dataset
Ds = pd.read_csv("data_cleaned_2021.csv")
```

```python
#Check if they have any missing value in data
Ds.isna().sum()
#Since the NA function said the data does not have any missing values, we manually decided the missing values.
```

```python
# Initialize an empty dictionary to store the missing value percentages
missing_percentage = {}

# Loop through each column in the DataFrame
for column in Ds.columns:
    # Count occurrences of '-1', 'unknown', and 'na' in the current column
    count_minus_1 = (Ds[column] == "-1").sum()  # Count '-1' values
    count_unknown = (Ds[column].str.lower() == "unknown").sum() if Ds[column].dtype == "object" else 0  # Count 'unknown' values (only for object dtype columns)
    count_na = (Ds[column].str.lower() == "na").sum() if Ds[column].dtype == "object" else 0  # Count 'na' values (only for object dtype columns)

    # Calculate the total number of values in the current column
    total_values = len(Ds[column])

    # Calculate the percentage of '-1', 'unknown', and 'na' values in the column
    percentage_minus_1 = (count_minus_1 / total_values) * 100  
    percentage_unknown = (count_unknown / total_values) * 100  
    percentage_na = (count_na / total_values) * 100 

    # Store the results for the current column in the dictionary
    missing_percentage[column] = {
        'count_minus_1': count_minus_1, 
        'percentage_minus_1': percentage_minus_1,  
        'count_unknown': count_unknown,  
        'percentage_unknown': percentage_unknown,  
        'count_na': count_na, 
        'percentage_na': percentage_na  
    }

# Print the missing value statistics for each column
for column, values in missing_percentage.items():
    print(f"Column: {column}")  
    print(f"  Count of '-1': {values['count_minus_1']}, Percentage of '-1': {values['percentage_minus_1']:.2f}%")  
    print(f"  Count of 'unknown': {values['count_unknown']}, Percentage of 'unknown': {values['percentage_unknown']:.2f}%")  
    print(f"  Count of 'na': {values['count_na']}, Percentage of 'na': {values['percentage_na']:.2f}%")  
    print("-" * 50)  
```

```python
# Impute missing values in the 'Headquarters' column by replacing '-1' with 'Unknown'
Ds['Headquarters'].replace([-1], ["Unknown"], inplace=True)

# Impute missing values in the 'Industry' column by replacing '-1' with 'Unknown'
Ds['Industry'].replace([-1], ["Unknown"], inplace=True)

# Impute missing values in the 'Sector' column by replacing '-1' with 'Unknown'
Ds['Sector'].replace([-1], ["Unknown"], inplace=True)

# Drop the 'job_title_sim' and 'seniority_by_title' columns from the DataFrame since they contain too much missing data or are not necessary for the analysis
Ds.drop(['job_title_sim', 'seniority_by_title'], axis=1, inplace=True)
```

```python
#Check the original dataset "Job title" values
Ds['Job Title'].unique()
```

```python
# Standardize and consolidate job title categories into unified labels
new = Ds['Job Title'] = Ds['Job Title'].apply(lambda x: 'Data Scientist' if 'scientist' in str(x).lower() else 
                                        'Data Engineer' if 'engineer' in str(x).lower() else
                                        'Data Analyst' if 'analyst' in str(x).lower() else
                                        'Director' if 'director' in str(x).lower() else
                                        'Manager' if 'manager' in str(x).lower() else
                                        'Consultant' if 'consultant' in str(x).lower() else
                                        'Modeler' if 'modeler' in str(x).lower() else
                                        'Other')
# View the result
new.unique()
```

```python
#combined location and job location, since these two is same content
Ds['Combined_Location'] = Ds['Location'].astype(str) + ' ' + Ds['Job Location'].astype(str)

# Remove redundancy if the 'Job Location' is the same as the last part of 'Location'
new_location = Ds['new_location'] = Ds['Combined_Location'].apply(lambda x: x if x.split()[-1] != x.split()[-2] else ' '.join(x.split()[:-1]))

# View the result
new_location
```

```python
# View the result
print(Ds.columns)
```

```python
# Function to clean and calculate the average salary from the 'Salary Estimate' column
def e_calculate_mean(salary_estimate_range):
    # Use regular expression to extract numeric values (removes "$" and "K")
    numbers = re.findall(r'\d+', salary_estimate_range)  
    
    # If two numbers (lower and upper bounds) are found in the salary range
    if len(numbers) == 2:
        # Convert the numbers to integers and calculate the mean salary
        lower, upper = int(numbers[0]), int(numbers[1])
        return (lower + upper) / 2  
    else:
        return None

# Apply the 'e_calculate_mean' function to the 'Salary Estimate' column to clean and calculate the mean
Ds['Avg_Salary(K)_Cleaned'] = Ds['Salary Estimate'].apply(e_calculate_mean)

# Calculate the quartile boundaries using qcut
# 'qcut' divides the data into 4 equal parts and returns the bin edges
quartile_boundaries = pd.qcut(Ds['Avg_Salary(K)_Cleaned'], 4, retbins=True)[1]

# Define labels for each quartile based on the calculated quartile boundaries
quartile_labels = [
    f"{round(quartile_boundaries[0], 1)}K - {round(quartile_boundaries[1], 1)}K",  
    f"{round(quartile_boundaries[1], 1)}K - {round(quartile_boundaries[2], 1)}K",  
    f"{round(quartile_boundaries[2], 1)}K - {round(quartile_boundaries[3], 1)}K",  
    f"{round(quartile_boundaries[3], 1)}K - {round(quartile_boundaries[4], 1)}K"   
]

# Use qcut again to assign the new labels based on the quartile boundaries
Ds['Salary_Quartile'] = pd.qcut(Ds['Avg_Salary(K)_Cleaned'], 4, labels=quartile_labels)

# Print the updated DataFrame to check the new 'Avg_Salary(K)_Cleaned' and 'Salary_Quartile' columns
print(Ds[['Avg_Salary(K)_Cleaned', 'Salary_Quartile']].head())
```

```python
# Drop the unnecessary columns that are no longer needed after the merge and cleaning process
# These columns either contain redundant information or are not relevant for further analysis
Ds_new = Ds.drop(['Location', 'Job Location', 'Competitors', 'Salary Estimate', 'Salary Estimate'], axis=1)

# Print the updated columns of the DataFrame after dropping unnecessary ones
print(Ds_new.columns)
```
