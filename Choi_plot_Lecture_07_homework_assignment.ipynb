# Bar plot with error bars
### For Q1-Q7, use the pyplot

```python
import matplotlib.pyplot as plt
```

```python
heights = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
errors = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]
```

### 1. Create a figure of width 10 and height 8.

```python
#set up the figure size
plt.figure(figsize=(10, 8))
#make a line plot
plt.plot()
#show the plot
plt.show()
```

### 2. Plot the blue bars, which have the heights listed in heights.

```python
#set up the bar range for heights number from heights array
plt.bar(range(len(heights)), heights)
plt.show()
```

### 3. Add error bars of cap size 5 and heights corresponding to the list error. (Hint: use [the yerr option](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html))

```python
# Add the error bars with cap size 5 on heights bar
plt.bar(range(len(heights)), heights, yerr=errors, capsize=5)


plt.show()
```

### 4. Set the axis to go from -0.5 to 6.5 on the x-axis and 70 to 95 on the y-axis.

```python
plt.bar(range(len(heights)), heights, yerr=errors, capsize=5)
#Set the axis min = -0.5, max= 6.5
plt.xlim(-0.5, 6.5)
#Set the yxis min = 70, max= 95
plt.ylim(70, 95)

plt.show()
```

### 5. Create an ax object using plt.subplot(). Use ax to set the x-axis ticks to be range(len(years)) and the x-axis labels to be the years list.

```python
# create an ax object subplot
fig, ax = plt.subplots()
plt.bar(range(len(heights)), heights, yerr=errors, capsize=5 )
plt.xlim(-0.5, 6.5)
plt.ylim(70, 95)
#set the range 7 which is the same number that counts years
ax.set_xticks(range(7))
#set up the label to years
ax.set_xticklabels(years)

plt.show()
```

### 6. Add the title "Final Exam Averages", x-axis label "Year", and y-axis label "Test average".

```python
fig, ax = plt.subplots()
plt.bar(range(len(heights)), heights, yerr=errors, capsize=5 )
plt.xlim(-0.5, 6.5)
plt.ylim(70, 95)
ax.set_xticks(range(7))
ax.set_xticklabels(years)
#set up the label x-axis
ax.set_xlabel('Year')
#set up the label y-axis
ax.set_ylabel('Test average')
#set up the title 
ax.set_title('Final Exam Average')


plt.show()
```

### 8. Create a similar figure created in Q6 using the seaborn library without the error bars.

```python
#panda for build the data frame 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

#build the data frame for use seaborn without error
info = pd.DataFrame([[82, 2000],
                     [84, 2001],
                     [83, 2002],
                     [86, 2003],
                     [74, 2004],
                     [84, 2005],
                     [90, 2006]],
                     columns=['Heights', 'Years'])

#build the bar plot x is years y is heights from info data
g = sns.catplot(x="Years", y="Heights", kind="bar", data=info)
#x min and max
g.ax.set_xlim(-0.5, 6.5)
#y min and max
g.ax.set_ylim(70, 95)
# x label
g.ax.set_xlabel('Year')
# y labe
g.ax.set_ylabel('Test average')
# title
g.ax.set_title('Final Exam Average')


```

### 9. Create a similar figure created in Q6 using the seaborn library with the error bars.

```python
#panda for build the data frame 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

#build the data frame for use seaborn with error
info = pd.DataFrame([[82, 2000,1.5],
                     [84, 2001,2.1],
                     [83, 2002,1.2],
                     [86, 2003,3.2],
                     [74, 2004,2.3],
                     [84, 2005,1.7],
                     [90, 2006,2.4]],
                     columns=['Heights', 'Years', 'errors'])

#build the bar plot x is years y is heights error bar from error 
g = sns.catplot(x="Years", y="Heights", yerr= info['errors'], kind="bar", data=info)

#x min and max
g.ax.set_xlim(-0.5, 6.5)
#y min and max
g.ax.set_ylim(70, 95)
# x label
g.ax.set_xlabel('Year')
# y label
g.ax.set_ylabel('Test average')
# title
g.ax.set_title('Final Exam Average')
```
